```python
import pandas as pd
import re

input_file = "/content/sunspots_data_1957-2023.csv"
output_file = "/content/sunspots_data_cleaned.csv"


"""
So basically I dropped all the unnecessary columns (shown below) and I also sorted by sunspot group
AND date and time. So it should go like: for each sunspot group number, record all of their Zurich classes
sorted by date and time. Then move on to the next sunspot group number, and so on...
"""



df = pd.read_csv(input_file, sep=";")
df.columns = df.columns.str.strip().str.strip("'")


# These are all the columns I don't think we need
cols_to_drop = [
    "observer",
    "seeing (SIDC)",
    "spot count",
    "weighted spot count",
    "latitude",
    "longitude",
    "manual latitude",
    "manual longitude",
    "ETH signature",
    "URL",
]

cols_to_drop = [c for c in cols_to_drop if c in df.columns]
df = df.drop(columns=cols_to_drop)


# Some of the group nr. rows had sub-numbers attached??? Like "245a" for eg., I dropped all of those.
group_col = "group nr."
df = df[df[group_col].astype(str).str.match(r"^\d+$")]

# Wanna sort by group nr, so convert it to int
df[group_col] = df[group_col].astype(int)

# Wanna also sort by group nr. AND time
sort_cols = [group_col, "date"]
if "time" in df.columns:
    sort_cols.append("time")

df = df.sort_values(by=sort_cols).reset_index(drop=True)


# Save pls
df.to_csv(output_file, index=False)
print(f"\nSUCCESS!!! {output_file}")
```
